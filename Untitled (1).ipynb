```python
Q1 = np.percentile(df['LotArea'], 25, interpolation='midpoint')
Q3 = np.percentile(df['LotArea'], 75, interpolation='midpoint')
```

```python
IQR = Q3-Q1
```

```python
lowerbound =Q1-1.5*IQR
upperbound =Q3+1.5*IQR
```

```python
df = df[(df.LotArea < upperbound) & (df.LotArea > lowerbound)]
```

```python
X = df.drop(columns = 'SalePrice')
Y= df['SalePrice']
```

```python
X
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
```

```python
X[:] = scaler.transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.2, random_state =42)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression

# Impute missing values in X_train
imputer = SimpleImputer(strategy='mean')  # You can use 'median' or 'most_frequent' as well
X_train_imputed = imputer.fit_transform(X_train)

# Initialize and fit the model
model = LinearRegression()
model.fit(X_train_imputed, Y_train)

# Impute missing values in X_test before making predictions
X_test_imputed = imputer.transform(X_test)
Y_pred = model.predict(X_test_imputed)
```

```python
Y_test[:5]
```

```python
Y_pred[:5]
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(Y_test,Y_pred)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Lasso

# Create an imputer object that will replace NaN values with the mean of the column
imputer = SimpleImputer(strategy='mean')

# Impute/fill the missing values in X_train
X_train_imputed = imputer.fit_transform(X_train)

# Initialize and fit the Lasso model
lasso_model = Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_model.fit(X_train_imputed, Y_train)

# Impute missing values in X_test
X_test_imputed = imputer.transform(X_test)

# Now you can make predictions
lasso_pred = lasso_model.predict(X_test_imputed)


```

```python
mean_absolute_error(Y_test, lasso_pred)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge

# Create an imputer object that will replace NaN values with the mean of the column
imputer = SimpleImputer(strategy='mean')

# Impute/fill the missing values in X_train
X_train_imputed = imputer.fit_transform(X_train)

# Initialize and fit the Ridge model
ridge_model = Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_model.fit(X_train_imputed, Y_train)

# If you are also using X_test, don't forget to impute the test data as well
X_test_imputed = imputer.transform(X_test)
ridge_pred = ridge_model.predict(X_test_imputed)


```

```python
mean_absolute_error(Y_test, ridge_pred)
    
```
